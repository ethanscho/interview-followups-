## K-means의 약점은 무엇일까?

### Answer first.
1. 구분할 cluster의 개수를 hyperparameter로 정해야만 한다. 
2. Circular cluster일 수밖에 없다.  
3. Intial grouping에 많은 영향을 받는다. 즉, 같은 데이터에 대해 같은 cluster로 나뉘리라는 보장이 없다.  
4. Outlier에 robust하지 않다.  


## Circular cluster이기 때문에 cluster를 잘 못 나누는 경우

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import time
import warnings

from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
n_samples = 500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.05)
```

```python
data = noisy_circles[0]
```

```python
MIN = -1.5
MAX = 1.5
```

```python
def plot(data):
    fig = plt.figure(figsize=(5, 5))

    colmap = {0: 'r', 1: 'g', 2: 'b'}

    for i, value in enumerate(data):
        plt.scatter(value[0], value[1], color=colmap[0], alpha=0.5)
        
#     for i, centroid in enumerate(centroids):
#         plt.scatter(centroid[0], centroid[1], color=colmap[i])

    plt.xlim(MIN, MAX)
    plt.ylim(MIN, MAX)
    plt.show()
```

```python
plot(data)
```

```python
K = 3
```

```python
centroids = [ np.random.ranf(2) * MAX for i in range(K)]

print(centroids)
```

```python
def assign():
    updated_clusters = np.zeros(n_samples)
    for data_idx, value in enumerate(data):
        min_distance = 1000.0
        cluster_idx = 0
        for i, centroid in enumerate(centroids):
            distance = (value[0] - centroid[0]) ** 2 + (value[1] - centroid[1]) ** 2
            if distance < min_distance:
                min_distance = distance
                cluster_idx = i
        
        updated_clusters[data_idx] = cluster_idx
    
    return updated_clusters
```

```python
def visualize():
    fig = plt.figure(figsize=(5, 5))

    colmap = {0: 'r', 1: 'g', 2: 'b'}

    for i, value in enumerate(data):
        plt.scatter(value[0], value[1], color=colmap[clusters[i]], alpha=0.5)
        
    for i, centroid in enumerate(centroids):
        plt.scatter(centroid[0], centroid[1], color=colmap[i])

    plt.xlim(MIN, MAX)
    plt.ylim(MIN, MAX)
    plt.show()
```

```python
def update_centroids():
    for i in range(K):
        indexes = clusters == i
        cluster = data[indexes]

        centroids[i][0] = np.mean(cluster[:,0])
        centroids[i][1] = np.mean(cluster[:,1])
```

```python
while True:
    update_centroids()
    updated_clusters = assign()
    
    if np.array_equal(updated_clusters, clusters) == True:
        break
        
    clusters = updated_clusters
```

```python
visualize()
```

군집이 서로 명확히 떨어져 있는 위와 같은 경우에도 circular cluster들로 나누어버림 

덧,
- initial grouping에 많은 영향을 받는 문제는 데이터 수를 늘려서 해결
- outlier에 취약한 문제는 centroid를 mean으로 계산하는 것이 아니라 median으로 구해서 해결 가능
